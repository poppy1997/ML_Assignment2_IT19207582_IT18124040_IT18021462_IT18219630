# DATA PREPROCESSING

```python
# pvc data type convert to int

ckd_df['pcv']=ckd_df['pcv'].astype(str).astype(np.int32)
```

```python
# Check the info

ckd_df.info()
```

```python
# convert classification val to 0 and 1

def claVal(x):
    if x == 'ckd':
        return 1
    if x == 'notckd':
        return 0
```

```python
ckd_df
```

```python
# convert rbc val to 0 and 1

def rbcVal(x):
    if x == 'normal':
        return 1
    if x == 'abnormal':
        return 0
```

```python
# Make new RBC column and apply rbcVal

ckd_df['RBC'] = ckd_df['rbc'].apply(rbcVal)
```

```python
# Drop the rbc column

ckd_df = ckd_df.drop(['rbc'], axis = 1)
```

```python
# convert pccc val to 0 and 1

def pccVal(x):
    if x == 'present':
        return 1
    if x == 'notpresent':
        return 0
```

```python
# Make new PCC column and apply pccVal

ckd_df['PCC'] = ckd_df['pcc'].apply(pccVal)
```

```python
# Drop the pcc column

ckd_df = ckd_df.drop(['pcc'], axis = 1)
```

```python
# convert ba val to 0 and 1
def baVal(x):
    if x == 'present':
        return 1
    if x == 'notpresent':
        return 0
```

```python
# Make new BA column and apply baVal

ckd_df['BA'] = ckd_df['ba'].apply(baVal)
```

```python
# Drop the ba column

ckd_df = ckd_df.drop(['ba'], axis = 1)
```

```python
ckd_df
```

```python
ckd_df['Classification'] = ckd_df['classification'].apply(claVal)
```

```python
ckd_df = ckd_df.drop(['classification'], axis = 1)
```

```python
ckd_df
```

# MODEL BUILDING

```python
# x= independent variable
x = ckd_df.values
# y = dependent variable
y = ckd_df['Classification'].values
```

```python
x = np.delete(x,15, axis = 1)
```

```python
#split the dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 0)
```

# Gradient Boost Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_clf = GradientBoostingClassifier()
gb_clf.fit(x_train, y_train)
gb_clf.score(x_test, y_test)
```

```python
#GB tuning
from sklearn import ensemble
gb_clf = ensemble.GradientBoostingClassifier(n_estimators = 40)
gb_clf.fit(x_train, y_train)
gb_clf.score(x_test, y_test)
```

```python
from sklearn import metrics
y_predic=gb_clf.predict(x_test)
print(metrics.classification_report(y_predic,y_test))
```

# Confussion Metrics

```python
from sklearn.metrics import confusion_matrix

cm=pd.DataFrame(
confusion_matrix(y_test, y_predic),columns=['Predicted:0', 'Predicted:1'],
index=['Actual:0', 'Actual:1']
)
ax= plt.subplot()
sb.heatmap(cm, annot=True, ax = ax)



ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix');
ax.xaxis.set_ticklabels(['0', '1']); ax.yaxis.set_ticklabels(['0', '1']);
```

# Random Forest Classifier

```python
rf_clf = ensemble.RandomForestClassifier(n_estimators = 100)
rf_clf.fit(x_train, y_train)
rf_clf.score(x_test, y_test)
```

```python
y_predic=rf_clf.predict(x_test)
print(metrics.classification_report(y_predic,y_test))
```

# Naive bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(x_train, y_train)
nb_clf.score(x_test, y_test)
```

```python
y_predic=nb_clf.predict(x_test)
print(metrics.classification_report(y_predic,y_test))
```

# KNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors = 3)
knn_clf.fit(x_train, y_train)
knn_clf.score(x_test, y_test)
```

```python
y_predic=knn_clf.predict(x_test)
print(metrics.classification_report(y_predic,y_test))
```

# Support Vector machine

```python
from sklearn.svm import SVC
svc_clf = SVC(probability = True, kernel = 'linear')
svc_clf.fit(x_train, y_train)
svc_clf.score(x_test, y_test)
```

```python
y_predic=svc_clf.predict(x_test)
print(metrics.classification_report(y_predic,y_test))
```

There are four most accurate classifiers are Randomforest,Gradient Boost and KNeighbors Classifiers. So that we select Random forest Classifier to train our model
